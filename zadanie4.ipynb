```
import requests
from bs4 import BeautifulSoup
import re

odpowiedz = requests.get("https://www.filmweb.pl/film/To%3A+Rozdzia≈Ç+2-2019-793838")
html_doc = odpowiedz.text
soup = BeautifulSoup(odpowiedz.text, 'html.parser')
rezyseria = soup.find(text="re≈ºyseria").next_element.find("a").text.strip()
premiera = soup.find(text="premiera").next_element.find("a").text.strip()
ocena = soup.find("span", itemprop="ratingValue").text.strip()
boxoffice = soup.find(text="boxoffice").next_element.text
print(f"ocena: {ocena}, rezyseria:{rezyseria}, premiera: {premiera}, boxoffice:{boxoffice}")
```
